```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType, TimestampType
```

```python
movie_language_schema = StructType(fields=[
    StructField("movieId", IntegerType(), False),
    StructField("languageId", StringType(), True),
    StructField("languageRoleId", StringType(), True)
])
```

```python
movie_language_df = spark.read \
    .schema(movie_language_schema) \
    .option("multiline", True) \
    .json(f"{bronze_folder_path}/{v_file_date}/movie_language")
```

## Parte 2 : Renombrar las columnas y a√±adir nuevas columnas


```python
from pyspark.sql.functions import col,concat,current_timestamp,lit

```

```python
movie_language_final_df = movie_language_df \
                                .withColumnsRenamed({"movieId": "movie_id",
                                                     "languageId": "lenguage_id"}) \
                                .drop("languageRoleId") \
                                .withColumn("ingestion_date", current_timestamp()) \
                                .withColumn("enviroment", lit("Production")) \
                                .withColumn("file_date", lit(v_file_date))




```
