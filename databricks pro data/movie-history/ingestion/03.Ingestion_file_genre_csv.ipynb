```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType
```

```python
genre_schema = StructType(fields=[
    StructField("genreId", IntegerType(), False),
    StructField("genreName", StringType(), True),
    ])
```

```python
genre_df = spark.read \
    .option("header", True)\
    .schema(genre_schema)\
    .csv(f"{bronze_folder_path}/{v_file_date}/genre.csv")
```

## Parte 2 : Seleccionar solo las columnas requeridas

```python
genre_selected_df = genre_df.select("genreId", "genreName")
```

```python
from pyspark.sql.functions import col
```

## Paso 3. Cambiar el nombre de las columnas segun lo requerido

```python
genre_rename_df = genre_selected_df \
    .withColumnRenamed("genreId", "genre_id") \
    .withColumnRenamed("genreName", "genre_name")


    
```

### Paso 4. Agregar la columna "Ingestion date" al dataframe

```python
from pyspark.sql.functions import current_timestamp, lit
# se importan funciones de fecha del dia de ingestion en el sistema y lit que nos sirve para agragar un valor en una columna ya que si se hace sin esta funcion no sera tomada como objeto

```

```python
genre_final_df = genre_rename_df \
    .withColumn("ingestion_date", current_timestamp())\
    .withColumn("enviroment", lit("production")) \
    .withColumn("file_date", lit(v_file_date))
```

### Paso 5. Escribir datos en el datalake en formato "Parquet"

```python
genre_final_df.write.mode("overwrite").format("delta").saveAsTable("movie_silver.genres")
```
