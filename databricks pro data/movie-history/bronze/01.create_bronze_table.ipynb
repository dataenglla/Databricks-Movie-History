## Crear tablas para archivos csv


#### Crear tablas movie

```sql
drop table if exists movie_bronze.movies;
create table if not exists movie_bronze.movies(
  movieId int,
  title string,
  budget double,
  homePage string,
  overview string,
  popularity double,
  yearReleaseDate int,
  releaseDate date,
  revenue double,
  durationTime int,
  movieStatus string,
  tagline string,
  voteAverage double,
  voteCount int
)
using csv
options (path "/mnt/moviehistory7809/bronze/movie.csv", header true)
```

```sql
select * from movie_bronze.movies;
```

```sql
drop table if exists movie_bronze.languages;
create table if not exists movie_bronze.languages(
  languageId int,
  languageCode string,
  languageName string
)
using csv
options (path "/mnt/moviehistory7809/bronze/language.csv", header true)
```

```sql
select * from movie_bronze.languages;
```

```sql
drop table if exists movie_bronze.genres;
create table if not exists movie_bronze.genres(
  genreId int,
  genreName string
)
using csv
options (path "/mnt/moviehistory7809/bronze/genre.csv", header true)
```

```sql
select * from movie_bronze.genres
```

```sql
drop table if exists movie_bronze.countries;
create table if not exists movie_bronze.countries(
  countryId int,
  countryIsoCode string,
  countryName string
)
using json options(path "/mnt/moviehistory7809/bronze/country.json")
```
